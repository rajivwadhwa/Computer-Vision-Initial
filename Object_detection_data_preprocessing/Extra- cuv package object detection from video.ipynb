```python
#Object detection from video 
#THis small code is used to detect coco classes for now
#we can make our own custom classes and give it under classes using this package
#Based on YOLOv5 TensorflowRT at back

pip install cvu-python

from vidsz.opencv import Reader, Writer
from cvu.detector import Detector

# set video reader and writer, you can also use normal OpenCV
reader = Reader("example.mp4")
writer = Writer(reader, name="output.mp4")

# create detector
detector = Detector(classes="coco", backend="tensorrt")

# process frames
for frame in reader:

    # make predictions.
    preds = detector(frame)

    # draw it on frame
    preds.draw(frame)

    # write it to output
    writer.write(frame)

writer.release()
reader.release()
```
