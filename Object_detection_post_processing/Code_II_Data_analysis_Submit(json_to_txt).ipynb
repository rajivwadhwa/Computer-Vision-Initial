```python
import numpy as np
import pandas as pd
import os
from google.colab import drive
drive.mount("/content/gdrive")
os.chdir("/content/gdrive/My Drive/YOLO_dataset")

#Importing Json file into dataframe in order to read the results

df = pd.read_json('/content/gdrive/My Drive/YOLO_dataset/GTSDB.json')
df.head()
```

```python
# Cleaning the dataframe to our specific required data from entire json file

abc = df['output'].iloc[0]
type(abc)
df_new = pd.DataFrame(abc) 
x = (pd.DataFrame())

df_new
```

```python
#The results in Sign column is our useful which is given in list, converting that list into dataframe

for index, row in df_new.iterrows():
  
    temp = row['signs']
    #print(temp)
    if(len(temp) != 0):
         temp[0]['frame_number'] = row['frame_number']
    else:
      temp.append({'frame_number':row['frame_number']})

    temp_df = pd.DataFrame(temp) 
    x = x.append(temp_df)


x
```

```python
#Rearranging the columns to get the same format as ground truth

x = x[['frame_number', 'coordinates', 'class', 'detection_confidence']]
x.to_csv (r'/content/gdrive/My Drive/YOLO_dataset/x.txt', index = False, sep = ';',header=False)
```
